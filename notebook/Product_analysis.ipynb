# Product Analysis

## 1. Best-Selling Products

We begin by identifying the best-selling products based on total sales. This analysis helps us understand which products are the most popular among customers.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('../data/cleaned_superstore.csv')

# Group by Product.Name and sum the sales
best_selling_products = df.groupby('Product.Name')['Sales'].sum().sort_values(ascending=False)

# Display the top 10 best-selling products
best_selling_products.head(10)
```

## 2. Product Profitability
Next, we calculate the total profit for each product to find out which products are the most profitable.

```python
# Group by Product.Name and sum the profits
product_profitability = df.groupby('Product.Name')['Profit'].sum().sort_values(ascending=False)

# Display the top 10 most profitable products
product_profitability.head(10)
```

## 3. Profit vs Sales Comparison
We will explore how the relationship between sales and profit looks for each product, which will help us understand if high sales correlate with high profits.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of Sales vs Profit, colored by Category
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category', alpha=0.7)
plt.title('Sales vs Profit by Product')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

## 4. Sales by Product Category
We analyze total sales by category to see which categories contribute the most to the overall sales.

```python
# Group by Category and sum the sales
category_sales = df.groupby('Category')['Sales'].sum()

# Sort the categories by sales and plot
category_sales.sort_values(ascending=False).plot(kind='bar', figsize=(10, 6), title='Sales by Category')
plt.ylabel('Total Sales')
plt.show()
```

## 5. Visualizing Product Performance
We visualize the sales and profitability of the top products using bar plots to easily compare their performance.

```python
# Top 10 best-selling products
top_products = df.groupby('Product.Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', figsize=(10, 6), title='Top 10 Best-Selling Products')
plt.ylabel('Sales')
plt.show()

# Top 10 most profitable products
top_product_profit = df.groupby('Product.Name')['Profit'].sum().sort_values(ascending=False).head(10)
top_product_profit.plot(kind='bar', figsize=(10, 6), title='Top 10 Most Profitable Products')
plt.ylabel('Profit')
plt.show()
```

## 6. Correlation Between Sales and Profit
Finally, we compute the correlation between sales and profit to understand how strongly these two variables are related for the products.

```python
# Calculate the correlation between Sales and Profit
correlation = df[['Sales', 'Profit']].corr()

# Plot the correlation matrix
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Sales and Profit')
plt.show()
```

## Conclusion
In this product analysis, we identified the best-selling and most profitable products, visualized their performance, and explored the correlation between sales and profit. This analysis is crucial for understanding which products contribute the most to your business and which ones may require further attention.
